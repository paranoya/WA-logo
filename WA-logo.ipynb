```python
%matplotlib ipympl
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.close('WA-logo')
fig = plt.figure('WA-logo', figsize=(10, 6))
axes = fig.subplots(nrows=1, ncols=1, squeeze=False,
                    sharex=True, sharey=True,
                    gridspec_kw={'bottom': .1, 'left': .05, 'top': .99, 'right': .99, 'hspace': 0, 'wspace': .1}
                   )
ax = axes[0, 0]
ax.set_xlim(-12, 20)
ax.set_ylim(-9, 9)
ax.set_aspect('equal')

line_width = 8

def plot_W(x0, y0, height, width):
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y = np.cos(x)
    ax.plot(x0+x*width/4/np.pi, y0+y*height/2, 'b-', lw=line_width)
    ax.plot(x0+x*width/4/np.pi, y0+y*height/2, 'w-', lw=.8*line_width)

def plot_A(x0, y0, height, width):
    x = np.array([-1, 0, 1])
    y = np.array([-1, 1, -1])
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)
    ax.plot(x0+x*width/2, y0+y*height/2, 'w-', lw=.8*line_width)
    ax.scatter(x0, y0+.5, color='w', edgecolor='b', s=100)
    x = np.linspace(-1, 1, 100)
    y = -1.5 + x**2/4
    ax.plot(x0+1.2*x*width/2, y0+y*height/2, 'w-')#, lw=line_width)

def plot_v(x0, y0, height, width, slant=0):
    x = np.array([-1., 0, 1])
    y = np.array([1, -1, 1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)

def plot_e(x0, y0, height, width, slant=0):
    x = np.array([1., -1, -1, 1])
    y = np.array([1, 1, -1, -1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)
    ax.scatter(x0-.5, y0, color='b', s=100)

def plot_p(x0, y0, height, width, slant=0):
    x = np.array([-.5, 1, 1, -1, -1])
    y = np.array([0, 0, 1, 1, -1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)

def plot_r(x0, y0, height, width, slant=0):
    x = np.array([1., -.5, 1, 1, -1, -1])
    y = np.array([-1, 0, 0, 1, 1, -1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)

def plot_t(x0, y0, height, width, slant=0):
    x = np.array([-1., 1])
    y = np.array([1, 1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)
    x = np.array([0., 0])
    y = np.array([-1, 1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)

def plot_i(x0, y0, height, width, slant=0):
    x = np.array([0., 0])
    y = np.array([-1, 1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)

def plot_f(x0, y0, height, width, slant=0):
    x = np.array([1., -1, -1])
    y = np.array([1, 1, -1])
    x += y*np.tan(slant*np.pi/180)
    ax.plot(x0+x*width/2, y0+y*height/2, 'b-', lw=line_width)
    ax.scatter(x0-.5, y0, color='b', s=100)


ax.scatter(0, 0, marker='H', color='b', alpha=.2, s=60000)
x0 = 0
y0 = 1.5
slant = 15
plot_W(x0-8, y0, 2, 5)
plot_e(x0-3, y0, 1, 2, slant)
plot_A(x0, 0, 4, 4)
plot_v(x0+3, y0, 1, 2, slant)
plot_e(x0+6, y0, 1, 2, slant)

plot_p(x0+4, -y0, 1, 2, -slant)
plot_e(x0+7, -y0, 1, 2, -slant)
plot_r(x0+10, -y0, 1, 2, -slant)
plot_t(x0+13, -y0, 1, 2, -slant)
plot_i(x0+15, -y0, 1, 1, -slant*1.5)
plot_f(x0+17, -y0, 1, 2, -slant)

ax.axis('off')
#ax.grid(alpha=.5)
plt.savefig('WA-logo.png')
plt.savefig('WA-logo.pdf')
plt.show()
```
